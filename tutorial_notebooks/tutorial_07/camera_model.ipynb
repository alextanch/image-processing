##### **Преобразование из декартовых координат в однородные**

Преобразование декартовых координат точек на плоскости в трехмерные точки 
с однородными координатами
$$
    \mathbf{x} =
    \begin{bmatrix}
        x \\
        y
    \end{bmatrix}
    \rightarrow
    \lambda\cdot
    \begin{bmatrix}
        x \\
        y \\
        1
    \end{bmatrix} =
    \widetilde{\mathbf{x}}
    \qquad
    \lambda \neq 0
$$

Это избыточное представление в том смысле, 
что различные $\lambda$ соответствуют одной декартовой точке.

При преобразовании декартовых координат в однородные, 
обычно выбирают $\lambda = 1$

$$
    \mathbf{x} =
    \begin{bmatrix}
        x \\
        y
    \end{bmatrix}
    \rightarrow
    \begin{bmatrix}
        x \\
        y \\
        1
    \end{bmatrix} =
    \widetilde{\mathbf{x}}
$$
    

##### **Преобразование из однородных координат в декартовы**

Обратное преобразование из однородных координат в декартовы:
$$
    \widetilde{\mathbf{x}} =
    \begin{bmatrix}
        \tilde{x} \\
        \tilde{y} \\
        \tilde{z}
    \end{bmatrix}
    \rightarrow
    \begin{bmatrix}
        \tilde{x}/\tilde{z} \\
        \tilde{y}/\tilde{z}
    \end{bmatrix} =
    \mathbf{x}
    \qquad
    \widetilde{z} \neq 0
$$

Точки с $\widetilde{z} = 0$ называются точками на бесконечности.

**Пример**:

$$
    \widetilde{\mathbf{x}} =
    \begin{bmatrix}
        2 \\
        3 \\
        1
    \end{bmatrix} = 
    \begin{bmatrix}
        4 \\
        6 \\
        2
    \end{bmatrix}
    \rightarrow 
    \begin{bmatrix}
        4/2 \\
        6/2
    \end{bmatrix} =
    \begin{bmatrix}
        2 \\
        3
    \end{bmatrix} =
    \mathbf{x}
$$

#### **Однородные координаты в пространстве**

Аналогично для трехмерного пространства: каждую точку в трехмерном пространстве 
$\mathbf{w}$ можно преобразовать в четырехмерную точку с однородными координатами 
$\widetilde{\mathbf{w}}$:
$$
    \mathbf{w} =
    \begin{bmatrix}
        u \\
        v \\
        w
    \end{bmatrix}
    \rightarrow
    \lambda\cdot
    \begin{bmatrix}
        u \\
        v \\
        w \\
        1
    \end{bmatrix} = 
    \widetilde{\mathbf{w}}
    \qquad \lambda \neq 0
$$

И обратно из однородных координат в декартовы:
$$
    \widetilde{\mathbf{w}}=
    \begin{bmatrix}
        \tilde{u} \\
        \tilde{v} \\
        \tilde{w} \\
        \tilde{z}
    \end{bmatrix}
    \rightarrow
    \begin{bmatrix}
        \tilde{u}/\tilde{z} \\
        \tilde{v}/\tilde{z} \\
        \tilde{w}/\tilde{z} \\
    \end{bmatrix}=
    \mathbf{w}
$$

Точки с $\tilde{z} = 0$ называются точками на бесконечности.
